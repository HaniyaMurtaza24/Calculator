```python
import tkinter as tk
from tkinter import *

cal = tk.Tk()
cal.title("Calculator")
operator = ""
text_input = StringVar()

def btnClick(numbers):
    global operator
    operator = operator + str(numbers)
    text_input.set(operator)

def btnEquals():
    global operator
    ans = str(eval(operator))
    text_input.set(ans)

def btnClear():
    global operator
    operator = ""
    text_input.set("")

txtDisplay = Entry(cal, font = ('arial', 20, 'bold'), textvariable = text_input, bd = 30, insertwidth = 4,
                   bg = "white", justify = "right")
txtDisplay.grid(columnspan = 4)

btn0 = Button(cal,padx = 16, bd = 8, fg = "crimson", font = ('arial', 20, 'bold'), text = "0",command= lambda:btnClick(0))
btn0.grid(row = 4, column = 1)

btn1 = Button(cal,padx = 16, bd = 8, fg = "crimson", font = ('arial', 20, 'bold'), text = "1",command= lambda:btnClick(1))
btn1.grid(row = 3, column = 0)

btn2 = Button(cal,padx = 16, bd = 8, fg = "crimson", font = ('arial', 20, 'bold'), text = "2",command= lambda:btnClick(2))
btn2.grid(row = 3, column = 1)

btn3 = Button(cal,padx = 16, bd = 8, fg = "crimson", font = ('arial', 20, 'bold'), text = "3",command= lambda:btnClick(3))
btn3.grid(row = 3, column = 2)

btn4 = Button(cal,padx = 16, bd = 8, fg = "crimson", font = ('arial', 20, 'bold'), text = "4",command= lambda:btnClick(4))
btn4.grid(row = 2, column = 0)

btn5 = Button(cal,padx = 16, bd = 8, fg = "crimson", font = ('arial', 20, 'bold'), text = "5",command= lambda:btnClick(5))
btn5.grid(row = 2, column = 1)

btn6 = Button(cal,padx = 16, bd = 8, fg = "crimson", font = ('arial', 20, 'bold'), text = "6",command= lambda:btnClick(6))
btn6.grid(row = 2, column = 2)

btn7 = Button(cal,padx = 16, bd = 8, fg = "crimson", font = ('arial', 20, 'bold'), text = "7",command= lambda:btnClick(7))
btn7.grid(row = 1, column = 0)

btn8 = Button(cal,padx = 16, bd = 8, fg = "crimson", font = ('arial', 20, 'bold'), text = "8",command= lambda:btnClick(8))
btn8.grid(row = 1, column = 1)

btn9 = Button(cal,padx = 16, bd = 8, fg = "crimson", font = ('arial', 20, 'bold'), text = "9",command= lambda:btnClick(9))
btn9.grid(row = 1, column = 2)

addition = Button(cal, padx = 16, bd = 8, fg = "crimson", font = ('arial', 20, 'bold'),text = "+",command= lambda:btnClick("+"))
addition.grid(row = 1, column = 3)

subraction = Button(cal, padx = 16, bd = 8, fg = "crimson", font = ('arial', 20, 'bold'),text = "-",command= lambda:btnClick("-"))
subraction.grid(row = 2, column = 3)

multiplication = Button(cal, padx = 16, bd = 8, fg = "crimson", font = ('arial', 20, 'bold'),text = "*",command= lambda:btnClick("*"))
multiplication.grid(row = 3, column = 3)

division = Button(cal, padx = 16, bd = 8, fg = "crimson", font = ('arial', 20, 'bold'),text = "/",command= lambda:btnClick("/"))
division.grid(row = 1, column = 3)

clear = Button(cal, padx = 16, bd = 8, fg = "crimson", font = ('arial', 20, 'bold'),text = "C",command= btnClear)
clear.grid(row = 4, column = 3)

equal = Button(cal, padx = 16, bd = 8, fg = "crimson", font = ('arial', 20, 'bold'),text = "=",command= btnEquals)
equal.grid(row = 4, column = 0)

cal.mainloop()
```
